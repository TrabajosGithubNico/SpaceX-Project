Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Crear una nueva columna marker_color basada en la columna class
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Mostrar las primeras filas para verificar
spacex_df.head()

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# for each row in spacex_df data frame
# create a Marker object with its coordinate
# and customize the Marker's icon property to indicate if this launch was successed or failed, 
# e.g., icon=folium.Icon(color='white', icon_color=row['marker_color']
for index, record in spacex_df.iterrows():
    # TODO: Create and add a Marker cluster to the site map
    # marker = folium.Marker(...)
    coordinate = [record['Lat'], record['Long']]
    
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(
            color='white',
            icon_color=record['marker_color'],  # verde o rojo
            icon='rocket',
            prefix='fa'
        ),
        popup=f"Launch Site: {record['Launch Site']}<br>Class: {record['class']}"
    )
    
    marker_cluster.add_child(marker)

site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

# Coordenadas del launch site (ajusta según el que estés usando)
launch_site_lat = 28.562302
launch_site_lon = -80.577356
launch_site_coordinate = [launch_site_lat, launch_site_lon]

# Coordenadas del punto de costa que seleccionaste con MousePosition
coastline_lat = 28.56367
coastline_lon = -80.57163
coastline_coordinate = [coastline_lat, coastline_lon]

# Calcular distancia
distance_coastline = calculate_distance(
    launch_site_lat, launch_site_lon,
    coastline_lat, coastline_lon
)

# 1. Crear marcador en el punto de costa
coastline_marker = folium.Marker(
    location=coastline_coordinate,
    popup="Closest Coastline Point",
    icon=folium.Icon(color='blue')
)
site_map.add_child(coastline_marker)

# 2. Dibujar la línea entre launch site y costa
folium.PolyLine(
    locations=[launch_site_coordinate, coastline_coordinate],
    weight=2,
    color='blue'
).add_to(site_map)

# 3. Mostrar la distancia como etiqueta usando DivIcon
distance_marker = folium.Marker(
    location=coastline_coordinate,
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_coastline)
    )
)
site_map.add_child(distance_marker)

site_map
```

Sí, muy cerca.

Conclusión:  
Los launch sites están intencionalmente ubicados cerca de ferrocarriles para facilitar logística industrial.

2. ¿Están los launch sites cerca de autopistas?

    Sí, también están muy cerca.

En el mapa, la distancia a la carretera principal suele ser entre 0.5 y 2 km.
Por qué es lógico

    Las autopistas permiten acceso rápido para personal, vehículos de soporte, transporte de equipos y emergencias.

    Un launch site necesita accesos terrestres robustos.

Conclusión:  
Los launch sites están estratégicamente conectados a autopistas para soporte operativo.

3. ¿Están los launch sites cerca de la costa?

Sí, extremadamente cerca.

En la medición, la distancia a la costa fue de 0.5–1.7 km, lo cual es típico.
Por qué es esencial

    Los lanzamientos se hacen hacia el océano para evitar que restos caigan sobre zonas habitadas.

    La costa permite rutas de lanzamiento seguras hacia el Atlántico.

    También facilita la recuperación de boosters y fairings.

Conclusión:  
La proximidad a la costa no es casual, es un requisito de seguridad y operación.

4. ¿Mantienen los launch sites cierta distancia respecto a ciudades?

    Seguridad: explosiones, ondas de choque, caída de escombros.

    Regulaciones federales exigen zonas de exclusión amplias.

Conclusión:  
Los launch sites están deliberadamente alejados de ciudades para minimizar riesgos y cumplir normativas.
